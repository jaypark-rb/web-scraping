## 더 이상 유효하지 않은 코드: Zara에서 화면 렌더링 방식을 변경함

```python
# 필요한 라이브러리 로드
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 스크랩 대상 URL(여기서는 자라 한국 사이트(여성 특가))
url = "https://www.zara.com/kr/ko/woman-special-prices-l1314.html"

# 설정할 헤더 정보
headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }

# 데이터 다운로드를 위해 GET 메소드로 요청
response = requests.request("GET", url, headers=headers)

# 다운로드된 데이터를 파싱(HTML 파서 활용)
data = BeautifulSoup(response.text, 'html.parser')
print(data)
```

```python
# 지정된 클래스 이름으로 모든 섹션을 탐색
cloths_data = data.find_all('div', attrs={'product-info _product-info '})

# 탐색된 상품 수
print('Total Number of Cloths Found : ', len(cloths_data))

# 상품에서 이름과 가격을 분리
for cloth in cloths_data:
    cloth_name = cloth.find('a')
    cloth_price = cloth.find('span', attrs={'sale'})
    if cloth_price:
        print(cloth_name.text, cloth_price['data-price'])
```
