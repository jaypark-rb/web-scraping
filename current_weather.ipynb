```python
# 필요한 라이브러리 로드
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 스크랩 대상 URL(여기서는 기상청)
url = "http://www.weather.go.kr/weather/observation/currentweather.jsp"

# 설정할 헤더 정보
headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }

# 데이터 다운로드를 위해 GET 메소드로 요청
response = requests.request("GET", url, headers=headers)

# 다운로드된 데이터를 파싱(HTML 파서 활용)
data = BeautifulSoup(response.text, 'html.parser')
print(data)
```

```python
# 지정된 클래스 이름으로 모든 섹션을 탐색
regions_tr = data.find_all('tr', class_=False)

# 탐색된 지역 수
print('Total Number of regions Found : ', len(regions_tr))

# 목록에서 개별 지역 분리
for regions_td in regions_tr:
    region_td = regions_td.find_all('td')
    # 지역명과 현재 기온을 출력
    print(region_td[0].text, region_td[5].text)
```
